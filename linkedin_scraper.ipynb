### Profile Web Scraping
using webscraping we can streamline our prompts and queries to include our experiences and expertise to make it easier to use the Groq Llama integration in the process of applying for internships and grad roles.

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.linkedin.com/in/biswas-simkhada/")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scraped text is from a LinkedIn profile. Your task is to extract and organize the following information into a structured JSON format:

    1. **Projects**: List of projects with the following keys: `project_name`, `description`, `github link`.
    2. **Volunteering**: Details of volunteering experiences with the following keys: `organization`, `role`, and `description`.

    Return the extracted information in a single JSON object with subsets for each of the categories mentioned above. Ensure that the JSON is valid and contains all extracted data accurately. Do not include any additional text or explanations; only provide the valid JSON object.
    ### VALID JSON (NO PREAMBLE):
    """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```
